OOP neden gerekli?
Karmaşık sistemleri yönetirken OOP kullanmak bize şu avantajları sağlar:

Düzen ve Gruplama: Kitap ve Üye gibi gerçek dünya varlıklarını kod içerisinde mantıklı objelere dönüştürür.

Sürdürülebilirlik: Bir özelliği (örneğin kitap adını) değiştirmek istediğimizde tüm kodu değil, sadece ilgili sınıfı güncellememiz yeterli olur.

Tekrar Kullanılabilirlik: Aynı kod yapısını binlerce farklı kitap veya üye için tekrar tekrar kullanabiliriz.

Kullanılan OOP Kavramları:
Classes & Objects (Sınıflar ve Nesneler): Book, Member ve DigitalBook şablonları oluşturulmuş ve bunlardan canlı örnekler (nesneler) türetilmiştir.

Encapsulation (Kapsülleme): Veriler (başlık, yazar, yıl) ve bu verilerle işlem yapan metotlar (get_info, borrow_book) aynı yapı içinde birleştirilmiştir.

Inheritance (Kalıtım): DigitalBook sınıfı, Book sınıfının tüm özelliklerini miras almış ve üzerine kendine has "dosya boyutu" ve "indirme" özelliklerini eklemiştir.

```python
class Book():
    def __init__(self,title,author,year,is_available=True):
        self.title= title
        self.author=author
        self.year= year
        self.is_available=is_available
    def get_info(self):
        print(f"Kitap Adı: {self.title}, Yazar: {self.author}, Yıl: {self.year}, Müsaitlik: {self.is_available}")
```

```python
class Member():
    def __init__(self,name,member_id):
        self.name=name
        self.member_id=member_id
    def borrow_book(self,book): #kitap müsaitse ödünç verilebilir.
        if book.is_available:
            book.is_available= False
            print(f"{self.name},'{book.title}' kitabını ödünç aldı.")
        else:
            print(f"Üzgünüz,{book.title} şuan başka birinde.")
    def return_book(self,book): # kitap iade edilirse müsait olduğunda.
        book.is_available=True
        print(f"{self.name}, '{book.title}' kitabını iade etti.")
```

```python
#Default değer kullanımı
kitap=Book("Nutuk","Mustafa Kemal Atatürk",1921)
kitap.get_info()
```

```python
#Inheritance
class DigitalBook(Book):
    def __init__ (self,title,author,year,file_size):
        self.title= title
        self.author=author
        self.year= year 
        self.file_size=file_size
    def download(self):
        print(f"Dosya indiriliyor...")
```

```python
#Sistem Testi
book=Book("Umberto Eco","Gülün Adı",1980)
book2=Book("Herge","Mavi Lotus",1936)
book3=DigitalBook("Diamond Suntra","Elmas Sutra",868,5)
member=Member("Esma",2211011)

member.borrow_book(book) #Esma, book1'i  ödünç aldı
member.borrow_book(book2) #Esma,book2'yi ödünç aldı
member.return_book(book) #Esma,book1'i teslim etti
book3.download() #Dijital book için download çağırdık.
```
